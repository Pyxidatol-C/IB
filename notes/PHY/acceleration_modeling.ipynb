```python
plt.scatter(t0 * t0, y0)
line = np.polyfit(t0 * t0, y0, 1).reshape(1, 2)
xs = np.linspace(0, 0.175).reshape(50, 1)
ys = np.concatenate((xs, np.ones((50, 1))), axis=1) @ line.T
plt.plot(xs, ys)
plt.xlim([0, 0.175])
plt.xlabel('$t^2$ (sec)')
plt.ylabel('displacement (cm)')
plt.legend([f"$y = {line[0, 0]} t^2 + {line[0, 1]}$"])

plt.figure()
plt.scatter(t0, y0)
xs_ = np.sqrt(xs)
ys_ = ys
plt.plot(xs_, ys_)
plt.xlabel('displacement (cm)')
plt.ylabel('t (sec)')
```
