# Motion
- Displacement ($m$): $x$ (initial displacement: $x_0$)
- Velocity ($m/s$): $v$ (initial velocity: $v_0$)
- Acceleration ($m/s^2$): $a$ 

## Graphs of motion
For constant acceleration with $v_0$ and $x_0$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 5)
a = t * 0 + 1
v = 1 + a * t
s = 1 + (v - 1) * t / 2

_, (a1, a2, a3) = plt.subplots(3, sharex=True)
for ax, y, y0_name, description in zip(
    (a1, a2, a3), 
    (a, v, s), 
    ('$a$', '$v_0$', '$x_0$'), 
    ('acceleration ($m/s^2$)', 'velocity ($m/s$)', 'displacement ($m$)')
):
    ax.plot(t, y)
    ax.grid(True, which='both')
    plt.xlabel('time ($s$)')
    ax.set_ylabel(description)
    plt.xticks(range(10))
    plt.xlim([0, 5])
    ax.set_ylim([0, max(y) + 1])
    ax.set_yticks([1])
    ax.set_yticklabels([y0_name])
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    ax.get_yaxis().set_label_coords(-0.1,0.5)
plt.tight_layout()
```

- $v = \frac{\Delta x}{\Delta t}$ (change in displacement over time)
- $a = \frac{\Delta{v}}{\Delta{t}}$ (change in velocity over time)

## Equation of motion for constant acceleration
$v = v_0 + at$  
$x = v_0 t + \frac{1}{2} a t^2 + x_0$  
$v^2 = {v_0}^2 + 2 a x$  
${v_y}^2 = v_{0y}^2 + 2 a y$

# Energy, Work and Power in mechanical systems

## Energy 
Energy is the potential to do work, measured in Joules.

### Kinetic energy
Kinetic energy is relative to motion:  
$$E_k = \frac{1}{2} m v^2$$


### Gravitational potential energy
The energy related to a gravitational field.  
$$E_p = m g h$$

- `m`: mass
- `g`: gravitational acceleration
- `h`: height

In a closed system with no heat losses, mechanical energy is conserved:  
$$E_p + E_k = \text{constant}$$

Energy is **always** conserved.


## Work
Work is the energy being transferred from one thing to another.

In mechanics, this transferrence is via a force.

$$W = F d$$

- `W`: $[J] = \frac{[kg] \times [m]^2}{[s]^2}$
- `F = m a`: $[N] = \frac{[kg] \times [m]}{[s]^2}$
- `d`(displacement) = $[m]$

Note: force and displacement should in the same direction ($W = cos \theta F s$)

## Power
rate at which energy transforms
$$P = \frac{\text{Energy}}{\text{Time}}$$
$$[Watt] = \frac{[Joule]}{[second]}$$
$$\text{Efficiency} = \frac{\text{Useful Work}}{\text{Total energy input}}$$

# Angular momentum
$w = \frac{v}{r}$

# Energy
$$K. E = \frac{1}{2} m v^2$$
$$P.E = m g h$$

Elastic potential energy: energy stored through compression / elongation

Total energy is neither created nor destroyed; it changes form and is conserved.

## Example

### Ball free falling

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

a = -9.81
g = 9.81
m = 1
t = np.linspace(0, 5)
h = a * (t ** 2) / 2
v = a * t
k_e = m * v ** 2 / 2  # kinetic energy
p_e = m * g * h + np.max(k_e)  # potential energy (relative to the max energy)
t_e = p_e + k_e  # total energy

plt.plot(t, k_e)
plt.plot(t, p_e)
plt.plot(t, t_e)
plt.xlabel("time")
plt.ylabel("E")
plt.xticks([])
plt.yticks([])
plt.legend(['P.E', 'K.E', 'T.E']);
```

$$[W] = [N] \cdot \frac{[m]}{[c]} = \frac{[kg] [m]}{[s^2]} \cdot \frac{[m]}{[s]} = [kg] \times \frac{[m]^2}{[s]^3}$$

$$P = F \cdot \frac{\Delta d}{\Delta t} = F v$$

# Momentum, Impulse and Force time diagrams

## Momentum
Momentum $p = m v$.

Linear momentum: momentum in $\left(\begin{array}{c}x \\ y \\ z\end{array}\right)$ directions

$$\vec{P_x} = m \vec{v_x}$$
$$\vec{P_y} = m \vec{v_y}$$

In a closed system, linear momentum is conserved.

### Example
Car A meets car B.

- $m_A = 500 kg$
- $v_A = 30 m/s$
- $m_B = 600 kg$
- $v_B = 0 m/s$

After impact, $m_{A+B} = 1100 kg$

$v_{A+B} = \frac{p_{A+B}}{m_{A+B}} = \frac{p_A + p_B}{m_A + m_B} = \frac{500 \times 30 + 600 \times 0}{500 + 600} = 13.6 m/s$

## Impulse
$= \text{Force} \times \Delta \text{time} = \Delta p$  
$F_{net} = m a = m \frac{\Delta v}{\Delta t} = \frac{\Delta m v}{\Delta t} = \frac{\Delta p}{\Delta t}$

# Elastic and inelastic collisions
- Elastic collision: objects collide and "bounce off" each other. Kinectic energy is conserved.
- Inelastic collisions: objects collide and "stick together". Kinectic energy is **not** conserved.

Momentum is conserved in both.

## Example
2 marbles:

- 1 mass 50g, still
- 1 mass 100g, v = 2m/s

### a) Find initial momentum 
$p = p_1 + p_2 = m_1 v_1 + m_2 v_2 = 0.050 \times 0 + 0.100 \times 2 = 0.2 Ns$

### b) Find initial kinetic energy
$E_k = E_{k1} + E_{k2} = \frac{1}{2} m_1 {v_1}^2 + \frac{1}{2} m_2 {v_2}^2 = 0.050 \times 0^2 + 0.100 \times 2^2 = 0.2 J$

### c) Find velocity of each marble after impact
$p = p_1 + p_2 = m_1 v_1 + m_2 v_2 = 0.2 Fs$  
$E_k = E_{k1} + E_{k2} = \frac{1}{2} (m_1 {v_1}^2 + m_2 {v_2}^2) = 0.2 J$

$v_1 = 2.67 m/s$  
$v_2 = 0.67 m/s$  


### d) Find the impulse for each marble
$I_1 = \Delta p_1 = 2.67 \times 0.050 - 0 \times 0.050 = 0.133 Ns$  
$I_2 = \Delta p_2 = 0.67 \times 0.100 - 2 \times 0.100 = -0.133 Ns$

### e) Find the force if the impact is 0.1s
$F_{net} = \frac{\Delta p}{\Delta t} = \frac{0.133 - 0.133}{0.1} = 0 N$

### f) Find the force on each marble if the impact is 0.1s
$F_1 = \frac{\Delta p}{\Delta t} = 1.33 N$  
$F_2 = \frac{\Delta p}{\Delta t} = -1.33 N$

## Inelastic
$m_1 v_1 + m_2 v_2 = (m_1 + m_2) v_f$

### Example
2 marbles from the last example stick together. $v_f$?

$v_f = \frac{(m_1 v_1 + m_2 v_2)}{m_1 + m_2} = \frac{0.2}{0.150} = 1.33 m/s$
